```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
flights = sns.load_dataset('flights')
flights
```

```python
flights =  flights.pivot_table(index= 'month', columns= 'year', values = 'passengers')
flights
```

```python
flights= flights.corr(numeric_only= True)
sns.heatmap(flights, cmap= 'inferno', annot= True, linewidth= 1, linecolor='black')
flights
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(flights, cmap= 'YlOrBr', linewidth=1, linecolor= 'black')
plt.show()
```

```python
sns.clustermap(flights,cmap= 'inferno', standard_scale=1, linewidth= 1, linecolor='gold', annot=True)
```
