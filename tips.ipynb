```python
import matplotlib.pyplot as plt
import  seaborn as sns 
```

```python
tip_df= sns.load_dataset('tips')
print(tip_df.to_string)
```

```python
sns.histplot(tip_df['total_bill'], bins= 20)
plt.xlabel('Total bill')
```

```python
sns.jointplot(x = 'total_bill', y = 'size', data= tip_df, kind= 'kde', fill=True)
```

```python
sns.pairplot(tip_df, hue ='sex', palette= 'inferno', diag_kind= 'hist')
```

```python
sns.barplot(x='sex', y='total_bill', data=tip_df)
```

```python
sns.countplot( x='sex', data= tip_df)
```

```python
sns.boxplot(x='day', y='total_bill',data=tip_df, hue='sex')
plt.legend(loc='upper left')
```

```python
plt.figure(figsize=(8,5))
sns.set_style('whitegrid')
sns.stripplot(x='day', y= 'total_bill', data=tip_df, jitter= True, hue='sex', dodge=True, palette='magma')
plt.ylabel('total bill')
```
