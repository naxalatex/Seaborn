```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ir_df= sns.load_dataset('iris')
print(ir_df.to_string)
```

```python
sns.histplot(ir_df['sepal_length'],  bins= 20)
```

```python
sns.jointplot(x='sepal_width', y='sepal_length', data= ir_df, kind='reg')
plt.xlabel('Sepal width')
plt.ylabel('Sepal length')
```

```python
sns.pairplot(ir_df, hue='species' , palette='inferno', diag_kind= 'hist')
```
